```python
import pandas as pd
import numpy as np
from statistics import median

%load_ext lab_black
```

```python
# загрузим данные
df_center = pd.read_csv("/home/sk27/repo/eremeev/test_projects/call_center_data.csv")
```

```python
# преобразуем формат данных о дате и времени совершения операции
df_center["datetime"] = df_center["date"] + " " + df_center["time"]
df_center["datetime"] = df_center["datetime"].astype("datetime64[ns]")
df_center.drop(["date", "time"], axis=1, inplace=True)
```

```python
# создадим пустой dataframe для хранения результатов
df_res = pd.DataFrame()

# перебираем в цикле действия каждого юзера
for unit in df_center.user_id.unique():
    # пустой dataframe для хранения результатов по юзеру
    df_result = pd.DataFrame()
    # фильтр по юзеру
    df_test = df_center[df_center["user_id"] == unit].reset_index(drop=True)
    # сортировка полученного набора по полю datetime (возрастание)
    df_test = df_test.sort_values("datetime").reset_index(drop=True)
    # для каждого действия определяем его продолжительность как интервал от момента его окончания
    # до момента окончания следующего действия
    for _ in range(1, df_test.shape[0]):
        delta = (
            df_test.loc[_, "datetime"] - df_test.loc[_ - 1, "datetime"]
        ).total_seconds()
        # сохраняем идентификатор действия и его продолжительность при условии, что продолжительность
        # не превышает 1200 секунд, и добавляем результат в dataframe юзера
        if delta <= 1200:
            row = {"action": (df_test.loc[_ - 1, "action"]), "duration": delta}
            df_result = df_result.append(row, ignore_index=True)
    # объединяем данные по юзерам
    df_res = pd.concat([df_res, df_result]).reset_index(drop=True)
# приводим результат к виду, установленному ТЗ и определяем медианные значения продолжительности действий
df_res.action = df_res.action.astype("int")
df_res = df_res.groupby("action").median("duration")
```
