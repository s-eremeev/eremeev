```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

res_1 = dict() # пустой словарь для хранения результата
url = 'https://spareskin.ru/all/t-shirts/' 
requests.get(url)
pages = requests.get(url)
soup = BeautifulSoup(pages.text, 'lxml')
names_goods = soup.find_all('h2', class_='woocommerce-loop-product__title')
prices = soup.find_all('div', class_='rey-productLoop-footer')
dresses = soup.find_all('div', class_='rey-breadcrumbs-item')
dress = dresses[2]
for i in range(0, len(names_goods)):
    res_1[names_goods[i].text] = prices[i].text
res_1 = pd.DataFrame.from_dict(res_1, orient ='index').reset_index().rename(columns={'index': 'goods', 0: 'price'})
res_1['price'] = res_1['price'].apply(lambda x: x.split('₽')[0].split(' ')[2])
res_1['dress'] = dress.text
res_1.head()
```

### "Посмотреть всё"

## Для первой страницы

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

res = list()
res_all = list() # пустой список для хранения результата
url = 'https://spareskin.ru/shop' 
requests.get(url)
pages = requests.get(url)
soup = BeautifulSoup(pages.text, 'lxml')
names_goods = soup.find_all('h2', class_='woocommerce-loop-product__title')
prices = soup.find_all('div', class_='rey-productLoop-footer')
dresses = soup.find_all('div', class_='rey-productCategories')
for i in range(0, len(names_goods)):
    res = list()
    res.append(names_goods[i].text)
    res.append(prices[i].text)
    res.append(dresses[i].text)
    res_all.append(res)
res_all = pd.DataFrame.from_records(res_all).reset_index().drop('index', axis = 1).rename(columns={0: 'goods', 1: 'price', 2: 'category'})
res_all['price'] = res_all['price'].apply(lambda x: x.split('\n')[1]).apply(lambda x: x.replace(' ', ''))
res_all['price'] = res_all['price'].apply(lambda x: x.rsplit('₽')[-2])
res_all
```

## Для второй страницы

https://spareskin.ru/shop/page/2/

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
res = list()
res_all_2 = list() # пустой список для хранения результата
url = 'https://spareskin.ru/shop/page/2/' 
requests.get(url)
pages = requests.get(url)
soup = BeautifulSoup(pages.text, 'lxml')
names_goods = soup.find_all('h2', class_='woocommerce-loop-product__title')
prices = soup.find_all('div', class_='rey-productLoop-footer')
dresses = soup.find_all('div', class_='rey-productCategories')
for i in range(0, len(names_goods)):
    res = list()
    res.append(names_goods[i].text)
    res.append(prices[i].text)
    res.append(dresses[i].text)
    res_all_2.append(res)
res_all_2 = pd.DataFrame.from_records(res_all_2).reset_index().drop('index', axis = 1).rename(columns={0: 'goods', 1: 'price', 2: 'category'})
res_all_2['price'] = res_all_2['price'].apply(lambda x: x.split('\n')[1]).apply(lambda x: x.replace(' ', '')).apply(lambda x: x.replace('-', ''))
res_all_2['price'] = res_all_2['price'].apply(lambda x: x.rsplit('₽'))
try:
    res_all_2['price'] = res_all_2['price'].apply(lambda x: x[len(x)-2])
except IndexError:
    res_all_2['price'] = res_all_2['price'].apply(lambda x: x[0])
res_all_2['price'] = res_all_2['price'].apply(lambda x: x.replace('Подробнее', 'Нет в наличии'))
res_all_2
```

```python
res = pd.concat([res_all, res_all_2]).reset_index()
res = res.drop('index', axis = 1)
res
```
