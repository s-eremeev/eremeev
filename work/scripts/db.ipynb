## 1. Подготовка данных

```python
%%time

# вспомогательные функции

def oktmo(o):
    """
    функция дополняет код ОКТМО до 11 символов или до 10 символов без ведущего 0
    """
    if len(str(o)) == 1:
        l = 9
    else:
        l = 11-len(str(o))
    return int(str(o) + "0"*(l))
    
def visit_count(row):
    """
    функция определяет количество посещений согласно условиям
    """
    if row["id_purpose"] >= 3:
        res = row["code_usl"] * row["kol_usl"] * row["size"]
    else:
        res = row["sl_id"] * row["size"]
    if np.isnan(res):
        return "не определено"
    else:
        return res
    
# добавление в файл данных строки заголовков
# старая строка: "year;month;region;clinic;oid;depart;branch;doctor_specialty;doctor_profile;terms;doctor_id;diagnosis;purpose;service_id;usl_service_id;usl_doctor_specialty;usl_clinic;usl_depart;sl_id;code_usl;kol_usl;yet_sum\n"
with open("/home/sk27/work/med/data/source.csv", "r") as f:
    first_line = next(f).strip()
    if first_line[0:4] != "year":
        with open(
            "/home/sk27/work/med/data/source.csv", "r", encoding="ISO-8859-1"
        ) as original:
            data = original.read()
        with open("/home/sk27/work/med/data/source.csv", "w") as modified:
            modified.write(
                "year;month;region;oid_mo_frmo;oid_mo_oms;oid_division;oid_office;id_doctor_specialty;id_doctor_profile;id_terms;id_doctor;id_diagnosis;id_purpose;id_service;id_service_local;id_service_provider;id_mo_oms_service;oid_division_service;sl_id;code_usl;kol_usl;yet_sum\n"
                + data
            )
    else:
        pass

# загрузка файла
df_data = pd.read_csv(
    "/home/sk27/work/med/data/source.csv",
    sep=";",
    encoding="ISO-8859-1",
    low_memory=False,
)


# удаление кавычек
for i, col in enumerate(df_data.columns):
    df_data.iloc[:, i] = df_data.iloc[:, i].astype("str").str.replace('"', "")
# df_data = df_data.apply(pd.to_numeric, errors="ignore")

# замена ВСЕХ запятых точками
for i, col in enumerate(df_data.columns):
    df_data.iloc[:, i] = df_data.iloc[:, i].astype("str").str.replace(',', ".")
df_data = df_data.apply(pd.to_numeric, errors="ignore")

# вставка столбца ОКТМО
df_data.insert(3, "oktmo_p", df_data["region"].apply(oktmo))

# проверка загруженных данных

# а) проверка количества столбцов - их должно быть именно 23 вместе с добавленным мной столбцом ОКТМО
if df_data.shape[1] > 23:
    print(f"Количество полей в файле (их {df_data.shape[1]}) не соответствует форме отчёта, должно быть 23.")
else:
    pass

# б) проверка столбца terms на "только 3"
try:
    if df_data["id_terms"].mean() != 3:
        print(
            "Нарушено требование об указании в поле 'Условия оказания медпомощи' только значения '3'."
        )
except:
    print("Не проходит проверка на '3'")
```

```python
# если в годе есть какя-то хрень
df_data.year = df_data.year.replace(
    "Ã¯Â»Â¿ÃÂ¯ÃÂ»ÃÂ¿ÃÂÃÂ¯ÃÂÃÂ»ÃÂÃÂ¿2022", "2023"
)
```

```python
# заменим все экселевско-csvшные nan на ортодоксальный пандосовский NaN
df_data = df_data.replace("nan", np.NaN)
```

## 2. Определяем количество посещений

```python
%%time

s2 = df_data.drop_duplicates().shape[0]

# подчёт, сколько раз строка повторяется в наборе
# при этом набор данных "свёртывается", каждая срока повторяется только один раз, но в столбце size отображается количество повторов
df_data = df_data.groupby(df_data.columns.tolist(), as_index=False).size()
print(
    f"Количество уникальных строк в исходном наборе данных: {s2}, количество строк после 'свёртывания' набора: {df_data.shape[0]}"
)

# определение количества посещений
df_data["visit_count"] = df_data.apply(visit_count, axis=1)
```

```python
df_data
```

```python
# количество строк в наборе
df_data.shape[0]
```

```python
# количество посещений
visit_count_sum = int(df_data.visit_count.sum())
```

```python
# повторим каждую строку столько раз, сколько было посещений
# если всё верно, количество строк должно стать равным количеству посещений
df_data_long = df_data.loc[df_data.index.repeat(df_data.visit_count)].reset_index(
    drop=True
)
```

```python
df_data_long
```

```python
# проверка
print(f"Количество строк в новом наборе: {df_data_long.shape[0]}")
print(f"Количество посещений: {visit_count_sum}")
if visit_count_sum == df_data_long.shape[0]:
    print("Всё верно!")
```

```python
df_data_long.columns
```

```python
# извлекаем из набора данные для формирования таблицы Расписание и индексируем строки
df_params = (
    df_data_long[
        [
            "oktmo_p",
            "oid_mo_oms",
            "oid_mo_frmo",
            "id_terms",
            "oid_division",
            "oid_office",
            "id_doctor",
            "id_doctor_specialty",
            "id_doctor_profile",
        ]
    ]
    .drop_duplicates()
    .reset_index(drop=True)
    .reset_index()
)
```

```python
# переименуем столбец index для последующего добавления максимального значения id, существующего в базе
df_params = df_params.rename(columns={"index": "bar"})
df_params
```

Теперь самое интересное: нужно: а) начать нумерацию с максимального значения уже существующего индекса +1; б) те строки, которые уже существуют, должны сохранить индекс прежним.

```python
# максимальное значение уже существующего в таблице schedule id
n = cur.execute("select max(s.id_schedule) from schedule s;").fetchall()
if pd.DataFrame(n).isnull().values.any():
    n = 0
else:
    n = n[0][0]
```

```python
n
```

```python
# создание нового столбца index, в котром нуменация продолжается относительно максимального значения, уже существующего в базе
df_params.insert(0, "index", df_params.bar + n + 1)
df_params = df_params.drop(columns="bar")
# сохраним в переменную количество столбцов в таблице df_params
count_columns = len(df_params.columns)
```

```python
df_params
```

```python
# в переменную df_schedule_table выгружаем существующую таблицу из базы
schedule_oktmo = df_params.oktmo_p[0]
query = "select * from schedule s where s.oktmo_p = " + str(schedule_oktmo) + ";"
df_schedule_table = pd.DataFrame(
    cur.execute(query).fetchall(),
    columns=[
        "index",
        "oktmo_p",
        "oid_mo_oms",
        "oid_mo_frmo",
        "id_terms",
        "oid_division",
        "oid_office",
        "id_doctor",
        "id_doctor_specialty",
        "id_doctor_profile",
    ],
)
```

```python
# это то, что в базе уже есть
df_schedule_table
```

```python
# преобразование типов данных для успешного соединения таблиц
df_params[["oid_division", "oid_office"]] = df_params[
    ["oid_division", "oid_office"]
].astype(object)
```

```python
# соединение таблиц
df_params = df_params.merge(
    df_schedule_table,
    left_on=[
        "oktmo_p",
        "oid_mo_oms",
        "oid_mo_frmo",
        "id_terms",
        "oid_division",
        "oid_office",
        "id_doctor",
        "id_doctor_specialty",
        "id_doctor_profile",
    ],
    right_on=[
        "oktmo_p",
        "oid_mo_oms",
        "oid_mo_frmo",
        "id_terms",
        "oid_division",
        "oid_office",
        "id_doctor",
        "id_doctor_specialty",
        "id_doctor_profile",
    ],
    how="left",
)
```

```python
df_params
```

```python
# проверяем, добавились ли в таблицу новые столбцы:
# если имеющиеся в загружаемой таблице строки дублируют уже существующие в базе,
# тогда добавится один столбец - index_y
if len(df_params.columns) > count_columns:
    df_params.insert(
        0,
        "index",
        df_params.apply(
            lambda row: row.index_x if np.isnan(row.index_y) else row.index_y, axis=1
        ),
    )
    df_params.reset_index(drop=True)
```
