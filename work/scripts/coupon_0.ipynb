# Копия расчётов дашборда

```python
%%time

import pandas as pd

df_test = pd.read_csv('C:/work/files/error.csv', sep=';', dtype=object, header=None)
df_tech_error = df_test.loc[(df_test[4].isin(['1', '2', '5', '10', '12'])) & (df_test[11]=='3')]
df_tech_error = pd.DataFrame(df_tech_error.groupby(1)[5].count()).reset_index()
df_org_error = df_test.loc[(df_test[4]=='8') & (df_test[11]=='3')]
df_org_error = pd.DataFrame(df_org_error.groupby(1)[5].count()).reset_index()
df_regions = pd.DataFrame(df_test[1]).drop_duplicates().drop(index=0).reset_index(drop=True)
df_errors = df_regions.merge(df_tech_error, left_on=1, right_on=1, how='left').merge(df_org_error, left_on=1, right_on=1, how='left')
df_errors = df_errors.rename(columns={1: 'region_name', '5_x': 'tech_errors', '5_y': 'org_errors'})
df_errors = df_errors.fillna(0)
df_errors[['tech_errors', 'org_errors']] = df_errors[['tech_errors', 'org_errors']].astype('int')
df_test = pd.read_csv('C:/work/files/good.csv', sep=';', dtype=object, header=None)
df_suc = pd.DataFrame(df_test.loc[df_test[8]=='3'].groupby(2).count()[0]).reset_index()
df_full = df_errors.merge(df_suc, left_on='region_name', right_on=2, how='left').rename(columns={0: 'success'})[['region_name', 'success', 'tech_errors', 'org_errors']].fillna(0)
df_full['success'] = df_full['success'].astype('int')
df_full = df_full.sort_values('region_name').reset_index(drop=True)
print('\n\n\tКоличество успешных запписей, организационных и технических ошибок по субъектам\n')
print(df_full.to_string(index=True))
```
