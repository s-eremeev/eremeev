### Группировка и агрегация

```python
# список путей к файлам данных

# пустой список для хранения путей
path_data_list = []
# спиок папок с данными субьектов
path = "C:/Users/Admin/work/med/data/data_full"
dir_list = os.listdir(path)
for region in dir_list:
    region_path = path + "/" + region
    # список файлов и директорий в директории субъекта (чтобы было)
    file_list = os.listdir(region_path)
    data_path = region_path + "/data.csv"
    df_region_data = pd.read_csv(data_path, sep=";", index_col=0)
    df_region_data = df_region_data.apply(pd.to_numeric, errors="ignore")
    # количество строк в исходном наборе
    s2 = df_region_data.shape[0]
    # заменим все пустые значения на 0
    df_region_data = df_region_data.fillna(666999)
    # сгруппируем по всем строкам, кроме числа визитов, а число визитов проссумируем
    df_region_data_group = df_region_data.groupby(
        df_region_data.columns.tolist()[0:-2], as_index=False
    ).sum()
    # проверка
    print(f"Регион: {region}")
    print(
        f"Количество строк в исходном наборе {s2}, после группировки {df_region_data_group.shape[0]}, сокращение: {s2 - df_region_data_group.shape[0]}"
    )
    if df_region_data["visit_count"].sum() == df_region_data_group["visit_count"].sum():
        n = df_region_data_group["visit_count"].sum()
        # флаг неизменности
        test_visit_count = 0
        print(f"Количество визитов не изменилось: {n}")
    else:
        # флаг неизменности
        test_visit_count = 1
        print("Изменилось количество визитов!")
    # шруппировка по специальности
    df_spec = pd.DataFrame(
        df_region_data_group.groupby(["month", "id_doctor_specialty"])[
            "visit_count"
        ].sum()
    ).reset_index()
    # проверка
    if df_spec["visit_count"].sum() == n:
        print(
            f"Данные сгруппированы по специальности, сумма количества визитов {n}, не изменилось"
        )
    else:
        print("Данные сгруппированы по специальности, изменилось количество визитов")
    # группировка по профилю
    df_profile = pd.DataFrame(
        df_region_data_group.groupby(["month", "id_doctor_profile"])[
            "visit_count"
        ].sum()
    ).reset_index()
    if df_profile["visit_count"].sum() == n:
        print(
            f"Данные сгруппированы по профилю, сумма количества визитов {n}, не изменилось"
        )
    else:
        print("Данные сгруппированы по профилю, изменилось количество визитов")
    # группировка по цели посещения
    df_purpose = pd.DataFrame(
        df_region_data_group.groupby(["month", "id_purpose"])["visit_count"].sum()
    ).reset_index()
    if df_profile["visit_count"].sum() == n:
        print(
            f"Данные сгруппированы по цели посещения, сумма количества визитов {n}, не изменилось"
        )
    else:
        print("Данные сгруппированы по цели посещения, изменилось количество визитов")
    break
```

## 1. Подготовка данных

```python
%%time

# вспомогательные функции

def oktmo(o):
    """
    функция дополняет код ОКТМО до 11 символов или до 10 символов без ведущего 0
    """
    if len(str(o)) == 1:
        l = 9
    else:
        l = 11-len(str(o))
    return int(str(o) + "0"*(l))
    
def visit_count(row):
    """
    функция определяет количество посещений согласно условиям
    """
    if row["purpose"] >= 3:
        res = row["code_usl"] * row["kol_usl"] * row["size"]
    else:
        res = row["sl_id"] * row["size"]
    if np.isnan(res):
        return "не определено"
    else:
        return res
    
# добавление в файл данных строки заголовков
with open("/home/sk27/work/med/data/source.csv", "r") as f:
    first_line = next(f).strip()
    if first_line[0:4] != "year":
        with open(
            "/home/sk27/work/med/data/source.csv", "r", encoding="ISO-8859-1"
        ) as original:
            data = original.read()
        with open("/home/sk27/work/med/data/source.csv", "w") as modified:
            modified.write(
                "year;month;region;clinic;oid;depart;branch;doctor_specialty;doctor_profile;terms;doctor_id;diagnosis;purpose;service_id;usl_service_id;usl_doctor_specialty;usl_clinic;usl_depart;sl_id;code_usl;kol_usl;yet_sum\n"
                + data
            )
    else:
        pass

# загрузка файла
df_data = pd.read_csv(
    "/home/sk27/work/med/data/source.csv",
    sep=";",
    encoding="ISO-8859-1",
    low_memory=False,
)


# удаление кавычек
for i, col in enumerate(df_data.columns):
    df_data.iloc[:, i] = df_data.iloc[:, i].astype("str").str.replace('"', "")
# df_data = df_data.apply(pd.to_numeric, errors="ignore")

# замена ВСЕХ запятых точками
for i, col in enumerate(df_data.columns):
    df_data.iloc[:, i] = df_data.iloc[:, i].astype("str").str.replace(',', ".")
df_data = df_data.apply(pd.to_numeric, errors="ignore")

# вставка столбца ОКТМО
df_data.insert(3, "oktmo", df_data["region"].apply(oktmo))

# проверка загруженных данных

# а) проверка количества столбцов - их должно быть именно 23 вместе с добавленным мной столбцом ОКТМО
if df_data.shape[1] > 23:
    print(f"Количество полей в файле (их {df_data.shape[1]}) не соответствует форме отчёта, должно быть 23.")
else:
    pass

# б) проверка столбца terms на "только 3"
try:
    if df_data["terms"].mean() != 3:
        print(
            "Нарушено требование об указании в поле 'Условия оказания медпомощи' только значения '3'."
        )
except:
    print("Не проходит проверка на '3'")
```
