```python
str = "прп7688рптпр8   7"



def loweralpha(str):
    """функция удаляет из строки все не буквенно-цифровые символы
    и возвращает результат в нижнем регистре"""

    lst = []

    for letter in str:

        if letter.isalnum():
            lst += letter.lower()

    return "".join(lst)



loweralpha(str)
```
