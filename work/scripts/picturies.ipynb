```python
import sqlite3, os

# заливка картинок-шаблонов в db
img_path = "C:/work/new_bot/img/img_source/"
conn = sqlite3.connect(r"C:\work\new_bot\img\db_img.sqlite")
cur = conn.cursor()
query = """INSERT INTO images (id, image) VALUES (?, ?)"""

for img in os.listdir(img_path):
    id = img
    i_path = img_path+img
    with open(i_path, 'rb') as file:
        blob_data = file.read()
    data = (id, blob_data)
    cur.execute(query, data)
    conn.commit()
cur.close()

# теперь они там
```

```python
sub_path = "C:/work/new_bot/img/sub/"
rus_path = "C:/work/new_bot/img/rus/"
save_parh = "C:/work/new_bot/img/save/"

data_dict = {
    # "января": [31, 1],
    # "февраля": [29, 2],
    # "марта": [31, 3],
    # "апреля": [30, 4],
    # "мая": [31, 5],
    # "июня": [30, 6],
    # "июля": [31, 7],
    # "августа": [31, 8],
    "сентября": [30, 9],
    # "октября": [31, 10],
    # "ноября": [30, 11],
    # "декабря": [31, 12],
}

sub_dict = {
    "sub_0": ["m0d", "M0"],
    "sub_1": ["m1d", "M1"],
    "sub_2": ["m2d", "M2"],
    "sub_3": ["m3d", "M3"],
    "sub_4": ["m4d", "M4"],
}

rus_dict = {
    "rus_0": ["Запись к врачу_Динамика показателей_", "Запись к врачу_Динамика показателей"],
    "rus_1": ["Запись к врачу_Ситуация в субъектах_", "Запись к врачу_Ситуация в субъектах"],
    "rus_2": ["Запись к врачу_Статистика ФЭР по РФ_", "Запись к врачу_Статистика ФЭР по РФ"],
    "rus_3": ["Запись к врачу_Технические ошибки_", "Запись к врачу_Технические ошибки"],
}

year_four = "2024"
year_two = "24"

for key, lst in data_dict.items():
    month_name = key
    month_number = "0"+str(lst[1]) if lst[1]<10 else str(lst[1])
    for i in range(1, lst[0]+1):
        date_str = " ".join([str(i), month_name, year_four])
        day_number = "0"+str(i) if i<10 else str(i)
        for img in os.listdir(sub_path):
            img_path = "".join([sub_path, img])
            img_preffix = img.split(".")[0]
            img_file_name = "".join([sub_dict.get(img_preffix)[0], day_number, month_number, year_two, ".png"])
            save_path = "".join([save_parh, sub_dict.get(img_preffix)[1], "/", img_file_name])
            im = Image.open(img_path)
            draw_text = ImageDraw.Draw(im)
            draw_text.text((135, 880), date_str, fill=("#444444"), font=font_date)
            im.save(save_path)
            print(save_path)
            # break # папки
        for img in os.listdir(rus_path):
            img_path = "".join([rus_path, img])
            img_preffix = img.split(".")[0]
            img_file_name = "".join([rus_dict.get(img_preffix)[0], day_number, ".png"])
            save_path = "".join([save_parh, rus_dict.get(img_preffix)[1], "/", img_file_name])
            im = Image.open(img_path)
            draw_text = ImageDraw.Draw(im)
            draw_text.text((135, 880), date_str, fill=("#444444"), font=font_date)
            im.save(save_path)
            print(save_path)
            print(save_path)
            # break # папки
    break # месяц
```
