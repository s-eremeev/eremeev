# Создание пустых папок

```python
# %%time

import pandas as pd
import os

dir_path = os.getcwd().replace("\\", "/")
dir_ = os.listdir(dir_path)
input(
    "положите в директорию со скриптом список имён папок в формате .csv (только один!) и нажмите любую кнопку"
)
path = ""
for element in dir_:
    try:
        if element.split(".")[-1] == "csv":
            path = dir_path + "/" + element
            print(f"ваши папки будут созданы в директории {dir_path+'/'}")
        else:
            pass
    except Exception:
        pass
if path == "":
    print("нет подходящего файла с перечнем создаваемых папок, давай до свидания!")
else:
    sub_list = pd.read_csv(path, header=None)[0].values.tolist()
    for element in sub_list:
        try:
            os.mkdir(element)
            print(f"\t- создана папка '{element}'")
        except Exception:
            print(f"папка '{element}' в директории {dir_path+'/'} уже существует")
```

```python
import pandas as pd
import numpy as np
```

```python
df_test = pd.read_excel("C:/work/test/sverdl.xlsx", engine = "openpyxl").dropna(how="all").reset_index(drop=True)
```

```python
df_test["mo"] = np.nan
df_test["dept"] = np.nan
df_test["doctor"] = np.nan
```

```python
for index in range(0, df_test.shape[0]-1):
    value = df_test.iloc[index]['Оперативный отчёт по количеству занятых бирок и фактических посещений']
    if pd.isna(df_test.iloc[index]['Оперативный отчёт по количеству занятых бирок и фактических посещений']) and pd.isna(df_test.iloc[index+1]['Оперативный отчёт по количеству занятых бирок и фактических посещений']):
        df_test.iloc[index]["doctor"] = value
    elif pd.isna(df_test.iloc[index+1]['Оперативный отчёт по количеству занятых бирок и фактических посещений']):
        df_test.iloc[index]["mo"] = value
    else:
        df_test.iloc[index]["dept"] = value
```

```python
df_test.shape[0]
df_test.columns

# заполняем пустые ячейки вниз
for index in range(0, df_test.shape[0]):
    for col in df_test.columns[0:3]:
        if pd.isna(df_test.iloc[index][col]):
            df_test.iloc[index][col] = df_test.iloc[index-1][col]
        else:
            pass

# заполняем пустые ячейки вверх
for index in range(0, 5):
    for col in df_test.columns:
        if pd.isna(df_test.iloc[index][col]):
            df_test.iloc[index][col] = df_test.iloc[index+1][col]
        else:
            pass
```
